```python
def parseVec(file, delimiter, word2int):
    lines = open(file, encoding='utf8').readlines()
    vocab_size, embed_size = [int(s) for s in lines[0].split()]
    vocab_size  = min(len(word2int), vocab_size)
    embeddings = np.zeros((vocab_size + 1, embed_size), dtype=np.float64)
    for i in range(1, vocab_size):
        try:
            line = lines[i][:-1].split(delimiter)
            word = line[0]
            if '*###*' in word or 'unk' in word:
                continue
            if word in word2int:
                wordvec = np.array([np.float64(j) for j in line[1:] if j != ''])
                embeddings[word2int[word]] = wordvec
        except Exception as e:
#             print(lines[i])
            print(e)
    return embeddings
```

```python
def display_closestwords_tsnescatterplot(model, words, embed_size):
    
    arr = np.empty((0, embed_size), dtype='f')
#     word_labels = [word]

    # get close words
#     close_words = model.similar_by_word(word)
#     print(arr.shape, model[word].shape)
    # add the vector for each of the closest words to the array
#     arr = np.append(arr, np.array([model[word]]), axis=0)
#     for wrd_score in close_words:
#         wrd_vector = model[wrd_score[0]]
#         word_labels.append(wrd_score[0])
#         arr = np.append(arr, np.array([wrd_vector]), axis=0)
    word_labels = words
    for word in words:
        wrd_vector = model[word]
        arr = np.append(arr, np.array([wrd_vector]), axis=0)
        
    # find tsne coords for 2 dimensions
    tsne = TSNE(n_components=2, random_state=0)
    np.set_printoptions(suppress=True)
    Y = tsne.fit_transform(arr)

    x_coords = Y[:, 0]
    y_coords = Y[:, 1]
    # display scatter plot
    plt.scatter(x_coords, y_coords)

    for label, x, y in zip(word_labels, x_coords, y_coords):
        plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
    plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
    plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
    
    plt.rc('font', family='Ebrima')
#     nx.draw(g, with_labels=True, font_family="Ebrima")
    plt.show()
```

```python
display_closestwords_tsnescatterplot(full_model, plot_words, embed_size=w2v.shape[1])
```
